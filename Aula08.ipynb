### Missão 01: Organizador de Arquivos.
Este script automatiza a organização de arquivos movendo-os para pastas nomeadas de acordo com suas extensões.

```python
import os
import shutil

# Define o caminho do diretório a ser organizado (pasta atual)
caminho = "/content/"

# Lista todos os itens no diretório
for arquivo in os.listdir(caminho):
    # Obtém o caminho completo do item
    caminho_completo = os.path.join(caminho, arquivo)

    # Verifica se é um arquivo (ignora pastas)
    if os.path.isfile(caminho_completo):
        # Extrai a extensão do arquivo e converte para minúsculo
        nome, extensao = os.path.splitext(arquivo)
        extensao = extensao.replace(".", "").lower()

        # Se o arquivo tiver uma extensão
        if extensao:
            # Define o nome da pasta de destino
            pasta_destino = os.path.join(caminho, extensao)

            # Cria a pasta se ela não existir
            if not os.path.exists(pasta_destino):
                os.makedirs(pasta_destino)
                print("Criada pasta para: " + extensao)

            # Move o arquivo para a pasta correspondente
            shutil.move(caminho_completo, os.path.join(pasta_destino, arquivo))
            print("Arquivo " + arquivo + " movido para a pasta " + extensao)
```
