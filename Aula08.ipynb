### Missão 01: Organizador de Arquivos.
Este script automatiza a organização de arquivos movendo-os para pastas nomeadas de acordo com suas extensões.

```python
import os
import shutil

# Define o caminho do diretório a ser organizado (pasta atual)
caminho = "/content/"

# Lista todos os itens no diretório
for arquivo in os.listdir(caminho):
    # Obtém o caminho completo do item
    caminho_completo = os.path.join(caminho, arquivo)

    # Verifica se é um arquivo (ignora pastas)
    if os.path.isfile(caminho_completo):
        # Extrai a extensão do arquivo e converte para minúsculo
        nome, extensao = os.path.splitext(arquivo)
        extensao = extensao.replace(".", "").lower()

        # Se o arquivo tiver uma extensão
        if extensao:
            # Define o nome da pasta de destino
            pasta_destino = os.path.join(caminho, extensao)

            # Cria a pasta se ela não existir
            if not os.path.exists(pasta_destino):
                os.makedirs(pasta_destino)
                print("Criada pasta para: " + extensao)

            # Move o arquivo para a pasta correspondente
            shutil.move(caminho_completo, os.path.join(pasta_destino, arquivo))
            print("Arquivo " + arquivo + " movido para a pasta " + extensao)
```

# Missão 02: Consulta simples a APIS

```python
import requests

# Solicita o CEP ao usuário
cep = input("Digite o CEP (apenas números): ")

# Define a URL da API ViaCep usando concatenação
url = "https://viacep.com.br/ws/" + cep + "/json/"

# Realiza a requisição GET
resposta = requests.get(url)

# Verifica se a requisição foi bem-sucedida
if resposta.status_code == 200:
    dados = resposta.json()

    # Verifica se o CEP não foi encontrado
    if "erro" not in dados:
        # Exibe as informações formatadas usando concatenação
        print("Endereço encontrado:")
        print("Logradouro: " + str(dados["logradouro"]))
        print("Bairro: " + str(dados["bairro"]))
        print("Cidade: " + str(dados["localidade"]))
        print("Estado: " + str(dados["uf"]))
    else:
        print("Erro: CEP não encontrado.")
else:
    print("Erro ao consultar a API. Status: " + str(resposta.status_code))
```

# Missão 03: Sistema de Notificação

```python
# Simula a leitura de um sensor de carga de processamento
carga_sistema = 95

# Define o limite crítico de segurança
limite_critico = 90

# Verifica se a condição crítica foi atingida
if carga_sistema > limite_critico:
    # Dispara a notificação via console simulando um alerta urgente
    print("--- ALERTA DE SISTEMA ---")
    print("NOTIFICAÇÃO: Carga crítica detectada em " + str(carga_sistema) + "%")
    print("Ação recomendada: Encerrar processos secundários imediatamente.")
else:
    # Caso esteja estável, apenas informa o status
    print("Monitoramento ativo. Status do sistema: Estável (" + str(carga_sistema) + "%)")
```
