# Número Primo:

**Antes - Ineficiente:**

```
def eh_Primo(n):
  for i in range(2, n):
    if n % i == 0:
      return False
  return True
```

**Depois - Eficiente:**

```
import math

def eh_Primo(n):
  if n < 2:
    return False
  # Otimização: verificar apenas até a raiz quadrada de n
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True

# Teste
numero = 17
print(f'O número {numero} é primo? {eh_Primo(numero)}')
```

#Inversão de String:

**Versão ineficiente:**

```
def invete_texto(texto):
  invertido = ''
  for letra in texto:
    invertido = letra + invertido
  return invertido
```

**Versão Eficiente**

```
def inverte_texto(texto):
  return texto[::-1]

# Teste
texto_original = 'Python'
print(f'Original: {texto_original}')
print(f'Invertido: {inverte_texto(texto_original)}')
```
